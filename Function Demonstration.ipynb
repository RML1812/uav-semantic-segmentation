```python
# Import function and class from .py file
from functions import *
from KMeansClass import KMeansModel
```

```python
import matplotlib.pyplot as plt
```

## Demonstration

```python
# Load dataset as image list and mask list (numpy)
images, masks = load_dataset()
```

```python
# Initialize KMeans Model
model = KMeansModel()
```

```python
# Train dataset: numpy_list, cluster_size, max_iter
model.fit(images, 6, 50)
```

```python
# Get a set of image and ground truth (mask)
image, mask = get_image_and_mask(images, masks)

f, axarr = plt.subplots(1,2)
axarr[0].imshow(image)
axarr[1].imshow(mask)
```

```python
# Use model to predict cluster
results = model.predict(image)

plt.imshow(results)
```

```python
# Evaluate image with ground truth (mask)
score = model.evaluate_single(image, mask)

print(f"V-Measure Score on set: {score}")
```

```python
# Evaluate image list with ground truth (mask) list
score = model.evaluate(images, masks)

print(f"V-Measure Score on Dataset: {score}")
```

```python
# Train dataset (without defining cluster_size): image_list, mask_list, max_iter. Returns best cluster size
cluster_size = model.fit_auto(images, masks, 50)

print(f"Best cluster size: {cluster_size}")
```

```python
# Use model to predict cluster
results = model.predict(image)

plt.imshow(results)
```

```python
# Evaluate image with ground truth (mask)
score = model.evaluate_single(image, mask)

print(f"V-Measure Score on set: {score}")
```

```python
# Evaluate image list with ground truth (mask) list
score = model.evaluate(images, masks)

print(f"V-Measure Score on Dataset: {score}")
```
